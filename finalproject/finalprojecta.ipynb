## Regression Model

```python
from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
df_hb.head()
```

```python
target_x = df_hb['Cumulative male cases'].head(10)
x = np.array([target_x])
x.reshape(-1, 1)
x
```

```python
target_y = df_hb['Cumulative female cases'].head(10)
y = np.array([target_y])
y
```

```python
#Creating a model with model based on ordinary least squares and fit it

model = LinearRegression().fit(x, y)
```

```python
#Checking whether the model works satisfactorily and interpret it

r_sq = model.score(x, y)
print('coefficient of determination:', r_sq)
```

```python
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

```python
#new_model = LinearRegression().fit(x, y.reshape((-1, 1)))
#new_model = LinearRegression().fit(x, y.reshape((-1, 1)))
#print('intercept:', new_model.intercept_)
#print('slope:', new_model.coef_)
```

```python
y_pred = model.predict(x)
print('predicted response:', y_pred, sep='\n')
```

```python
plt.scatter(x, y, alpha=0.5)
plt.plot(x, y_pred)
plt.title('Scatter plot x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
y_pred = model.intercept_ + model.coef_ * x
print('predicted response:', y_pred, sep='\n')
```
