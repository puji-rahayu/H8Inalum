## K-Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_breast_cancer
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
X = -2 * np.random.rand(100, 2)
X1 = 1 + 2 * np.random.rand(50, 2)

X[50:100, :] = X1
```

```python
print(X)
```

```python
print(X1)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
wcss = []

for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 10), wcss)
plt.title('The elbow method')
plt.xlabel('Number of cluster')
plt.ylabel('WCSS')
plt.show()
```

```python
Kmean = KMeans(n_clusters=2)
Kmean.fit(X)
```

```python
Kmean.cluster_centers_
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(-1.06215057, -0.91104922, s=200, marker='s')
plt.scatter(2.06075854,  1.87858447, s=200, marker='s')
plt.show()
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std=0.5, shuffle=True, random_state=0)
```

```python
plt.scatter(
    X[:, 0], X[:, 1],
    c='white', marker='o',
    edgecolors='black', s=50
)
plt.show()
```

```python
km = KMeans(n_clusters=3)
y_km = km.fit_predict(X)
```

```python
km.cluster_centers_
```

```python
plt.scatter(
    X[y_km == 0, 0], X[y_km == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolors='black',
    label='cluster 1'
)

plt.scatter(
    X[y_km == 1, 0], X[y_km == 1, 1],
    s=50, c='orange',
    marker='o', edgecolors='black',
    label='cluster 2'
)

plt.scatter(
    X[y_km == 2, 0], X[y_km == 2, 1],
    s=50, c='lightblue',
    marker='v', edgecolor='black',
    label='cluster 3'
)

# plot the centroids
plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)

plt.legend()
plt.grid()
plt.show()
```

```python
wcss = []

for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 10), wcss)
plt.title('The elbow method')
plt.xlabel('Number of cluster')
plt.ylabel('WCSS')
plt.show()
```

## PCA 

```python
from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
```

```python
breast_data = breast.data
```

```python
breast_data.shape
```

```python
breast_labels = breast.target
```

```python
breast_labels.shape
```

```python
labels = np.reshape(breast_labels, (569, 1))
```

```python
labels.shape
```

```python
final_breast_data = np.concatenate([breast_data, labels], axis=1)
```

```python
final_breast_data.shape
```

```python
import pandas as pd
```

```python
breast_datasets = pd.DataFrame(final_breast_data)
```

```python
breast_datasets.head()
```

```python
features = breast.feature_names
```

```python
print(features)
```

```python
features_labels = np.append(features, 'label')
```

```python
print(features_labels)
```

```python
breast_datasets.columns = features_labels
```

```python
breast_datasets.head()
```

```python
breast_datasets['label'].replace(0, 'Benign', inplace=True)
```

```python
breast_datasets['label'].replace(1, 'Malignant', inplace=True)
```

```python
breast_datasets.head()
```

```python
breast_datasets.tail()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = breast_datasets.loc[:, features].values
x = StandardScaler().fit_transform(x)
```

```python
x.shape
```

```python
np.std(x)
```

```python
fast_cols = ['features'+str(i) for i in range(x.shape[1])]
```

```python
print(fast_cols)
```

```python
normalised_breast = pd.DataFrame(x, columns=fast_cols)
normalised_breast.head()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principal_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_df = pd.DataFrame(data=principal_breast, columns=['PC 1', 'PC 2'])
```

```python
principal_breast_df.head()
```

```python
print('Variance Ratio : {}'.format(pca_breast.explained_variance_ratio_))
```

```python
100-(44.2+18.9)
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)

targets = ['Benign', 'Malignant']
colors = ['r', ' g']

for target, color in zip(targets, colors):
    indicesToKeep = breast_datasets['label'] == target
    plt.scatter(principal_breast_df.loc[indicesToKeep, 'PC 1'],
               principal_breast_df.loc[indicesToKeep, 'PC 2'], c = color, s = 50)
    
plt.legend(targets, prop={'size':15})
plt.show()
```
